```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import mglearn
from sklearn.model_selection import train_test_split
```

Linear regression aka ordinary least square.
OLS: finds the parameters that minimize the mean squared error between predictions and the true regression targets y on the training set. 
\begin{align*}
min\sum(y_i-\hat{y_i})^2
\end{align*}


```python
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

Ridge regression
\begin{align*}
 \sum(y_i-\hat{y_i})^2+\lambda\sum_{j=1}^p \beta_j^2
\end{align*}

we wanna the magnitude of coefficients to be as samll as possible 
Regularization means deliberately restricting a model to avoid overfitting, it means "fit the data well, but don't do anything extreme". 

* L2 regularization:
1. ‚úÖ Shrinks coefficients toward zero 
2. ‚úÖ Reduces variance
3. ‚úÖ Stabilizes solutions under multicollinearity
4. ‚úÖ Improves prediction performance

* But:
‚ùå It does NOT set coefficients exactly to zero
(that‚Äôs L1 regularization)



```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)
ridge.score(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
ridge10.score(X_train, y_train)
```

Lasso
\begin{align*}
 \sum(y_i-\hat{y_i})^2+\lambda \sum_{j=1}^p |\beta_j|
\end{align*}


we wanna the sum of the absolute values of the coefficients to be small. 
* L1 regularization:

1. ‚úÖ Shrinks coefficients toward zero
2. ‚úÖ Can set some coefficients exactly to zero
3. ‚úÖ Performs automatic feature selection
4. ‚úÖ Produces sparse models

* ‚ùó But:
‚ùå It can be unstable when predictors are highly correlated
(Unlike Ridge, it may pick one variable and ignore others.)

üîπ Key conceptual difference

* Ridge (L2) ‚Üí Shrinks all coefficients smoothly
* Lasso (L1) ‚Üí Shrinks and can eliminate coefficients

```python
from sklearn.linear_model import Lasso
lasso = Lasso().fit(X_train, y_train)
lasso.score(X_test, y_test)
```
