🌳 Gradient Boosting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import mglearn
from sklearn.datasets import make_blobs
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target,
                                                    random_state=0)
gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train, y_train)

print("accuracy on training set: %f" % gbrt.score(X_train, y_train))
print("accuracy on test data: %f" % gbrt.score(X_test, y_test))
```

```python
#max_depth

gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)
print("accuracy on training set: %f" % gbrt.score(X_train, y_train))
print("accuracy on test data: %f" % gbrt.score(X_test, y_test))
```

```python
#learning rate

gbrt = GradientBoostingClassifier(random_state=0, learning_rate=0.01)
gbrt.fit(X_train, y_train)
print("accuracy on training set: %f" % gbrt.score(X_train, y_train))
print("accuracy on test data: %f" % gbrt.score(X_test, y_test))
```

Both methods of decreasing the model complexity decreased the training set accuracy as expected. In this case, lowering the maximum depth of the trees provided a significant improvement of the model, while lowering the learning rate only increased the generalization performance slightly.

```python
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)

plt.plot(gbrt.feature_importances_, "o")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
```

1. Core idea

Gradient boosting builds an additive model in a stage-wise manner, where each new tree is trained to correct the errors (residuals) of the previous ensemble.

It perfomrs functional gradient descent in function space.

\begin{align*}
F_M(x) = \sum_{m=1}^{M}\gamma_mh_m(x)
\end{align*}

Where:
- $h_m(x)$=weak learner (usually shallow tree)
- $\gamma_m$ = step size
- M = number of boositng iterations

2. Intuition

Intead of:
Reduce variance by averaging many trees (Random Forest)

Boosting says:
Reduce bias by sequentially improving the model

Each new tree focuses on :

- Where the model performs poorly
- The direction of steepest loss reduction

3. Key Hyperparameters
- $n_estimators$ -> number of trees (M)
- learnning_reate -> step size
- max_depth -> tree depth (controls complexity)
- subsample -> stochastic boosting (optional)

Important tradeoff:

Small learning rate -> need more trees

Large learning reate -> risk overfitting

4. Bias-Variance perspective

- Random Forest -> mainly reduces variance
- Gradient Boosting -> mainly reduces bias

Boosting buils strong learners from weak learners

Because trees are shallow:
- individual tree = high bias
- Ensemble = low bias, controlled variance




| Random Forest   | Gradient Boosting |
| --------------- | ----------------- |
| Parallel trees  | Sequential trees  |
| Bagging         | Boosting          |
| Reduce variance | Reduce bias       |
| Less sensitive  | More sensitive    |
| Faster          | Slower            |
