# Decision Function vs Predicted Probability

This notebook clarifies the difference between:
- `decision_function`
- `predict_proba`

Both relate to model confidence, but they represent different layers of interpretation.


## 1. Conceptual Overview

Most binary classifiers internally compute a **real-valued score**:

$$ F(x) $$

This is called the **decision function value**.

From this score, we may derive a probability:

$$ p(x) = P(y=1 \mid x) $$

Pipeline view:

X → decision_function → link function (e.g., sigmoid) → predict_proba → threshold → predict


## 2. Decision Function

The decision function is the model's raw score:

- If F(x) > 0 → class 1
- If F(x) < 0 → class 0
- |F(x)| indicates confidence (distance from boundary)

Examples:

- Logistic Regression: F(x) = wᵀx + b (log-odds)
- SVM: signed margin distance
- Gradient Boosting (logistic loss): additive log-odds model


```python
# Example usage
# model.decision_function(X)

# Output shape (binary): (n_samples,)
# Output shape (multiclass): (n_samples, n_classes)
```

## 3. Predicted Probability

Probabilities are derived from the raw score.

For logistic-type models:

$$ p(x) = \frac{1}{1 + e^{-F(x)}} $$

Properties:
- Values in [0, 1]
- Each row sums to 1 (multiclass)
- Interpretable as calibrated uncertainty (if well-calibrated)


```python
# Example usage
# model.predict_proba(X)

# Output shape (binary): (n_samples, 2)
# Column 0 = P(class 0)
# Column 1 = P(class 1)
```

## 4. Key Differences

| Aspect | decision_function | predict_proba |
|----------|-------------------|---------------|
| Scale | Unbounded real numbers | Between 0 and 1 |
| Meaning | Raw model score | Interpretable probability |
| Used for | Margin visualization, ranking | Threshold tuning, uncertainty |
| Boundary | 0 | 0.5 |


## 5. Multiclass Prediction and np.argmax

In multiclass classification, predictions are often made using:

`np.argmax(scores, axis=1)`

This selects the class with the largest score (or probability).


```python
import numpy as np

probs = np.array([
    [0.1, 0.7, 0.2],
    [0.8, 0.1, 0.1],
    [0.2, 0.3, 0.5]
])

predicted_class = np.argmax(probs, axis=1)
predicted_class
```

## 6. np.max vs np.argmax

- `np.max` → returns the maximum value
- `np.argmax` → returns the index of the maximum value

Example:


```python
a = np.array([3, 8, 2, 5])

np.max(a)      # 8
np.argmax(a)   # 1 (index of 8)
```

## Final Summary

- The decision function F(x) is the model's raw score.
- Predicted probability is a transformed interpretation of F(x).
- Classification is determined by thresholding the score or probability.
- In multiclass problems, np.argmax selects the most confident class.
